```python
import numpy as np
import pandas as pd 
import matplotlib as plt
```

```python
dates = pd.date_range('2020-1-1',periods=30, freq='D')
data = range(40,100,2)
df = pd.DataFrame(data, index= dates, columns= ['value'])
df.head()
df.shape
```

### .rolling() method 

```python
df['rw30'] = df.rolling(window='30D').mean()
rw10 = df.rolling(window = '10D').agg(['mean', 'median', 'std'])

```

```python
df.plot(subplots=False)
# rw10.plot(subplots = True)
```
